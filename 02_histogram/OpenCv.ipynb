```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read image
imgPath = './assets/3_6.jpg'
imgOrigin = cv2.imread(imgPath)
```

```python
# histogram
imgYuv = cv2.cvtColor(imgOrigin, cv2.COLOR_BGR2YUV)
imgYuv[:, :, 0] = cv2.equalizeHist(imgYuv[:, :, 0])
imgEqCV = cv2.cvtColor(imgYuv, cv2.COLOR_YUV2BGR)
```

```python
plt.imshow(imgOrigin, 'gray'), plt.title('Origin Image')
plt.show()
plt.imshow(imgEqCV, 'gray'), plt.title('Equalised Image')
plt.show()
```

```python
hist, bin = np.histogram(imgOrigin[3].flatten(), 256, [0, 256])
plt.plot(hist, color='slategrey'), plt.title('Original Image Histogram')
plt.show()
hist, bin = np.histogram(imgEqCV[3].flatten(), 256, [0, 256])
plt.plot(hist, color='slategrey'), plt.title('Equalised Image Histogram')
plt.show()
```
